```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)
```

```python
df = pd.read_csv("../data/Phishing.csv")
df.head(10)
```

```python
result_df = df['Result'].value_counts().to_frame()

result_df = result_df.reset_index()
result_df.columns = ["Class","Num_Observations"]
result_df
```

```python
plt.figure()
result_df.plot(kind='bar',x='Class',y='Num_Observations')
```
